```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

```

```python

faces = np.array([1,2,3,4,5,6])
pdf = np.zeros(6)
for i in range(6):
    for j in range(6):
        min_ij = np.min([faces[i],faces[j]])
        pdf[min_ij-1] += 1
pdf = pdf/36
cdf = np.cumsum(pdf)
```

```python
N = 1000
d1 = np.random.choice(faces, size=N)
d2 = np.random.choice(faces, size=N)
min_roll = [ min([d1[i], d2[i]]) for i in range(N) ]

ecdf = np.zeros(6)
for i in range(6):
    ecdf[i] = (1/N)*np.sum( min_roll <= faces[i] )
```

```python
plt.scatter(faces,ecdf,label='ECDF')
plt.plot(faces,ecdf,label='ECDF')
plt.scatter(faces,cdf,label='CDF')
plt.plot(faces,cdf,label='CDF')
plt.xlabel("x")
plt.ylabel("Probability")
plt.legend(loc='lower right')
plt.title('Empirical and Theoretical CDF')
plt.show()
```

```python
faces = np.array([1,2,3,4,5,6])
N = 5000
d1 = np.random.choice(faces, size=N)
d2 = np.random.choice(faces, size=N)
d3 = np.random.choice(faces, size=N)
med_roll = [ int(np.median([d1[i], d2[i], d3[i]])) for i in range(N) ]

expected_value = np.sum(med_roll)/N
print('Expected value: ', expected_value)

ecdf = np.zeros(6)
single = np.zeros(6)
for i in range(6):
    ecdf[i] = (1/N)*np.sum( med_roll <= faces[i] )
    single[i] = (1/N)*np.sum( d1 <= faces[i] )

expected_value = np.sum(med_roll)/N
```

```python
plt.scatter(faces,ecdf,label='Median')
plt.plot(faces,ecdf,label='Median')
plt.scatter(faces,single,label='Single')
plt.plot(faces,single,label='Single')
plt.xlabel("x")
plt.ylabel("Probability")
plt.legend(loc='lower right')
plt.title('Median and Single Roll CDF')
plt.show()
```

```python


N = 50000
results = []
for i in range(N):
    go = True
    sum = 0
    while go is True:
        roll = np.random.choice(faces, size=1)
        sum += roll
        if roll < 6:
            go = False
            results.append(int(sum[0]))

results = pd.DataFrame(results)

print(results.describe())

sns.histplot(results)
plt.show()

sns.ecdfplot(results)
plt.show()
```

```python
df = pd.read_csv('./the_bootstrap/data/mammogram.csv')
```

```python
#question 1
print( pd.crosstab( df['treatment'], df['breast_cancer_death']) , '\n')
print( pd.crosstab( df['treatment'], df['breast_cancer_death'], margins=True, normalize=True) , '\n')

raw_treatment_effect = 44425/44925 - 44405/44910
print('Raw treatment effect: ', raw_treatment_effect)
```

```python
#question 2
df['survive'] = 0
df.loc[ df['breast_cancer_death'] == 'no', 'survive'] = 1

df_treat = df.loc[ df['treatment'] == 'mammogram', :]
df_control = df.loc[ df['treatment'] == 'control', :]

S = 5000
fx = []
for t in range(S):
    df_treat_t = df_treat.sample( df_treat.shape[0], axis=0, replace=True)
    df_control_t = df_control.sample( df_control.shape[0], axis=0, replace=True)
    #
    treatment_effect_t = np.mean(df_treat_t['survive']) - np.mean(df_control_t['survive'])
    fx.append(treatment_effect_t)

sns.kdeplot(fx)
plt.show()

sns.ecdfplot(fx)
plt.show()
```

```python
#ques 3
level = .99
lower_bound = np.quantile(fx, .005 )
upper_bound = np.quantile(fx, .995)

print('99% confidence interval: (', lower_bound, ', ', upper_bound, ')' )
```

 The analysis indicates that receiving a mammogram may not have a statistically significant impact on 25-year survival rates. This conclusion is based on a 99% confidence interval that includes zero, suggesting that the observed difference in survival rates between the mammogram and control groups could be due to chance.


It is crucial to acknowledge the limitations of this analysis. Several factors could influence the results, including:

- Selection bias: If individuals who chose to receive mammograms were generally healthier or had lower breast cancer risk, the treatment group might appear more successful than it actually is.
- Confounding factors: Other risk factors for breast cancer, not accounted for in the analysis, could impact survival rates regardless of mammogram use.
- Subgroup effects: While mammograms may not show a significant overall benefit, they might be highly effective for specific subgroups or high-risk individuals.
- quality of life: The analysis focuses solely on mortality, but mammograms could improve other aspects of patients' lives, such as reducing the need for aggressive treatments.


To gain a more comprehensive understanding, further research should address these points. For example, investigating whether mammograms were randomly assigned or if there were systematic differences between the treatment and control groups. Additionally, exploring the potential impact of mammograms on various subpopulations and considering quality-of-life outcomes could provide valuable insights.
